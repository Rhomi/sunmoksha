### Second Level Processing

```python
day_names={0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
```

```python
# Code to add hourly usage into daily usage

for key, user in users.items():
    max_consumption = user.apply(lambda x: max(x[1:25]), axis=1)
    min_consumption = user.apply(lambda x: min(x[1:25]), axis=1) 
    max_hour = user.iloc[:,1:25].idxmax(axis=1)
    mean_consumption = user.iloc[:,1:25].mean(axis=1)
    std_dev = user.iloc[:,1:25].std(axis=1)
    user['UnitsConsumed'] = user.iloc[:,1:25].sum(1)
    user['MaxConsumption'] = max_consumption
    user['MinConsumption'] = min_consumption
    user['MaxHour'] = max_hour
    user['MeanConsumption'] = mean_consumption
    user['StdDev'] = std_dev
    user['DayOfTheWeek'] = user['Date'].dt.dayofweek.map(day_names)
    user.drop(user.columns[1:25], axis = 1, inplace=True)
    user.to_csv(key+".csv", index=None)
```
